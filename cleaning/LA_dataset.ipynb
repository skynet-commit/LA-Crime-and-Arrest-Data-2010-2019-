```python
# Create a def function to convert dates (eg 1152019-01-01T00:00:00) to the correct format using Regex

import re

def fix_mmddyyyy_prefix(s):

    s = str(s)

    m = re.match(r'^(\d{7})[^T]*(T.*)$', s)
    if m:
        date = m.group(1)
        month = int(date[0])
        day = int(date[1:3])
        year = int(date[3:])
        time = m.group(2)
        
        return f"{year:04d}-{month:02d}-{day:02d}{time}"
    else:
        return s
```

```python
crime_df['Date Reported'] = crime_df['Date Reported'].apply(fix_mmddyyyy_prefix)
```

```python
# Imported PDF to extract data and covert to JSON. Created a def function using Regex to read the 4-digit codes + read the description

import pdfplumber
import json

pdf_path = "MO_CODES_Numerical_20180627.pdf"
output_path = "mo_codes.json"

codes = {}

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if not text:
            continue

        for line in text.split("\n"):
            # match: 4-digit code + description
            match = re.match(r"^(\d{4})\s+(.+)", line)
            if match:
                code, description = match.groups()
                codes[code] = description.strip()

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(codes, f, indent=2, ensure_ascii=False)
```

```python
# Open the JSON file in read mode

with open('mo_codes.json', 'r') as file:
    mo_codes = json.load(file)
```

```python
#Pass the values from Json we created to a new column in the crime.df

def str_to_lst_mo(row):
    if pd.isna(row):
        return None
    else:
        row = str(row).split(' ')
        results = []
        for i in row:
            i = i.strip()
            results.append(mo_codes.get(i, f"UNKNOWN_MO_{i}"))
        return results
    
#crime_df['MO Codes Description']= crime_df['MO Codes'].apply(str_to_lst_mo)

crime_df.insert(
    10,
    'MO Codes Description',
    crime_df['MO Codes'].apply(str_to_lst_mo)
)
```

### importing the files needed to create a csv that matches coordinates with zip codes

```python

import geopandas as gpd

areas = gpd.read_file("Census_ZIP_Code_Tabulation_Areas.shp", engine="pyogrio")
areas = areas.to_crs("EPSG:4326")

print(areas.head())
print(areas.columns)
```

```python
crimes = pd.read_csv("crimes_latlon.csv")

gdf_crimes = gpd.GeoDataFrame(
    crimes,
    geometry=gpd.points_from_xy(crimes.Longitude, crimes.Latitude),
    crs="EPSG:4326"
)
```

```python
areas = areas[["ZCTA5", "geometry"]]
```

```python
joined = gpd.sjoin(
    gdf_crimes,
    areas,
    how="left",
    predicate="within"
)
```

```python
final = joined[[
    "CrimeID",
    "Year",
    "Latitude",
    "Longitude",
    "ZCTA5"
]]
```

```python
#export zip files to csv
final.to_csv("crimes_with_zip.csv", index=False)
```
